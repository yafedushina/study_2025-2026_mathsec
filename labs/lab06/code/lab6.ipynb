```python
import math
import random
from math import gcd  

def pollard_rho(n):
    c = random.randint(1, n - 1) 
    a = b = random.randint(1, n - 1)

    def f(x):
        return (x * x + c) % n

    while True:
        a = f(a) 
        b = f(f(b))  
        d = gcd(abs(a - b), n)

        if d != 1 and d != n:
            return d 
        elif d == n:
            return "Делитель не найден"  

n = 1359331
result = pollard_rho(n)
print(f"Нетривиальный делитель числа {n}: {result}")
```
